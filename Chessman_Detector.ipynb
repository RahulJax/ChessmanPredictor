```python
import matplotlib.pyplot as plt
import numpy as np
import cv2

%matplotlib inline
```

```python
class_names = ['BISHOP', 'KING', 'KNIGHT', 'PAWN', 'QUEEN', 'ROOK']
```

```python
width = 96
height= 96
```

```python
from glob import glob
from keras import preprocessing

#the folder Sample_Images contain various folder of names Bishop , King, Knight, Pawn etc
#which contain the respective images of the data sets
# the images datasets are downloaded from 'https://www.kaggle.com/niteshfre/chessman-image-dataset'

#load images type_1
images_type_1 = []
for image_path in glob('./Sample _Images/Bishop/*.*'):
    image = preprocessing.image.load_img(image_path, target_size=(width, height))
    x = preprocessing.image.img_to_array(image)
    images_type_1.append(x)
    
#load images type_2
images_type_2 = []
for image_path in glob('./Sample _Images/King/*.*'):
    image = preprocessing.image.load_img(image_path, target_size=(width, height))
    x = preprocessing.image.img_to_array(image)
    images_type_2.append(x)
    
#load images type_3
images_type_3 = []
for image_path in glob('./Sample _Images/Knight/*.*'):
    image = preprocessing.image.load_img(image_path, target_size=(width, height))
    x = preprocessing.image.img_to_array(image)
    images_type_3.append(x)
    
#load images type_4
images_type_4 = []
for image_path in glob('./Sample _Images/Pawn/*.*'):
    image = preprocessing.image.load_img(image_path, target_size=(width, height))
    x = preprocessing.image.img_to_array(image)
    images_type_4.append(x)
    
#load images type_5
images_type_5 = []
for image_path in glob('./Sample _Images/Queen/*.*'):
    image = preprocessing.image.load_img(image_path, target_size=(width, height))
    x = preprocessing.image.img_to_array(image)
    images_type_5.append(x)
    
#load images type_6
images_type_6 = []
for image_path in glob('./Sample _Images/Rook/*.*'):
    image = preprocessing.image.load_img(image_path, target_size=(width, height))
    x = preprocessing.image.img_to_array(image)
    images_type_6.append(x)
```

```python
X_type_1 =np.array(images_type_1)
X_type_2 =np.array(images_type_2)
X_type_3 =np.array(images_type_3)
X_type_4 =np.array(images_type_4)
X_type_5 =np.array(images_type_5)
X_type_6 =np.array(images_type_6)
print(X_type_1.shape)
print(X_type_2.shape)
print(X_type_3.shape)
print(X_type_4.shape)
print(X_type_5.shape)
print(X_type_6.shape)
```

```python
X =np.concatenate((X_type_1, X_type_2),axis=0)

if len(X_type_3):
    X=np.concatenate((X,X_type_3),axis=0)
    
if len(X_type_4):
    X=np.concatenate((X,X_type_4),axis=0)
        
if len(X_type_5):
    X=np.concatenate((X,X_type_5),axis=0)
        
if len(X_type_6):
    X=np.concatenate((X,X_type_6),axis=0)
    
#scale the data to[0,1] values
X=X/255.

X.shape
```

```python
from keras.utils import to_categorical

y_type_1= [0 for item in enumerate(X_type_1)]
y_type_2= [1 for item in enumerate(X_type_2)]
y_type_3= [2 for item in enumerate(X_type_3)]
y_type_4= [3 for item in enumerate(X_type_4)]
y_type_5= [4 for item in enumerate(X_type_5)]
y_type_6= [5 for item in enumerate(X_type_6)]

y=np.concatenate((y_type_1,y_type_2),axis=0)

if len(y_type_3):
    y=np.concatenate((y,y_type_3),axis=0)
    
if len(y_type_4):
    y=np.concatenate((y,y_type_4),axis=0)

if len(y_type_5):
    y=np.concatenate((y,y_type_5),axis=0)
    
if len(y_type_6):
    y=np.concatenate((y,y_type_6),axis=0)
    
y= to_categorical(y, num_classes=len(class_names))

print(y.shape)
    
print(y)
```

```python
from keras.models import Sequential
from keras.layers.core import Activation,Dropout, Flatten, Dense
from keras.layers.convolutional import Convolution2D, MaxPooling2D
from keras.optimizers import Adam

#default parameters
conv_1=16
conv_1_drop=0.2
conv_2=32
conv_2_drop = 0.2
dense_1_n =1024
dense_1_drop=0.2
dense_2_n=512
dense_2_drop=0.2
lr=0.001

epochs= 30
batch_size=32
color_channels=3

def build_model(conv_1_drop=conv_1_drop, conv_2_drop=conv_2_drop,
                 dense_1_n=dense_1_n ,dense_1_drop=dense_1_drop ,
                 dense_2_n=dense_2_n ,dense_2_drop=dense_2_drop ,lr=lr):
    model=Sequential()
    #first convolution layer of 16 filters and 3x3 size
    model.add(Convolution2D(conv_1, (3,3),input_shape=(width,height,color_channels), activation='relu'))
    model.add(MaxPooling2D(pool_size=(2,2)))
    model.add(Dropout(conv_1_drop))
    
    #adding second convolution layer of 32 filters and of 3x3 matrix 
    #followed by Maxpooling(to decrease the size of the input keeping important features)
    #and dropout(to prevent overfeeding)
    model.add(Convolution2D(conv_2, (3,3), activation='relu'))
    model.add(MaxPooling2D(pool_size=(2,2)))
    model.add(Dropout(conv_2_drop))
    
    #converting into 1d array
    model.add(Flatten())
    
    #adding a fully connected neural network of 1024 neurals to the flattened data 
    model.add(Dense(dense_1_n, activation='relu'))
    model.add(Dropout(dense_1_drop))
    
    model.add(Dense(dense_2_n, activation='relu'))
    model.add(Dropout(dense_2_drop))
    
    #adding the output layer neuron containg same number of neuron as the number of class_names 
    #and softmax is used to give the percentage
    model.add(Dense(len(class_names),activation='softmax'))
    
    model.compile(loss='categorical_crossentropy',optimizer=Adam(lr=lr), metrics=['accuracy'])
    
    return model
    
```

```python
import numpy as np
np.random.seed(1) #for reproducibility

#model with base parameters
model = build_model()
model.summary()
```

```python
epochs=30
```

```python
model.fit(X,y, epochs=epochs)
```

```python
import time

#get the reference to the webcam
camera=cv2.VideoCapture(0)
camera_height= 500

while(True):
    _,frame=camera.read()
    
    frame=cv2.flip(frame,1)
    
    #rescaling camera output
    aspect = frame.shape[1] / float(frame.shape[0])
    res = int(aspect * camera_height) #landscape orientation
    frame = cv2.resize(frame, (res, camera_height))
    
    #add rectangle
    cv2.rectangle(frame, (180,75), (550,425), (0, 255,0), 2)
    
    #get ROI
    roi=frame[75+2: 425-2, 150+2 :500-2]
    
    #parse BGR to RGB
    roi=cv2.cvtColor(roi,cv2.COLOR_BGR2RGB)
    
    #resize
    roi=cv2.resize(roi, (width,height))
    
    #predict
    roi_X=np.expand_dims(roi,axis=0)
    
    predictions=model.predict(roi_X)
    type_1_pred,type_2_pred,type_3_pred,type_4_pred,type_5_pred,type_6_pred = predictions[0]
    
    #add text
    type_1_text='{}: {}%'.format(class_names[0],int(type_1_pred*100))
    cv2.putText(frame, type_1_text, (70,120),cv2.FONT_HERSHEY_SIMPLEX, 0.6, (240,240,240),2)
    
    #add text
    type_2_text='{}: {}%'.format(class_names[1],int(type_2_pred*100))
    cv2.putText(frame, type_2_text, (70,150),cv2.FONT_HERSHEY_SIMPLEX, 0.6, (240,240,240),2)
    
    #add text
    type_3_text='{}: {}%'.format(class_names[2],int(type_3_pred*100))
    cv2.putText(frame, type_3_text, (70,180),cv2.FONT_HERSHEY_SIMPLEX, 0.6, (240,240,240),2)
    
    #add text
    type_4_text='{}: {}%'.format(class_names[3],int(type_4_pred*100))
    cv2.putText(frame, type_4_text, (70,210),cv2.FONT_HERSHEY_SIMPLEX, 0.6, (240,240,240),2)
    
    #add text
    type_5_text='{}: {}%'.format(class_names[4],int(type_5_pred*100))
    cv2.putText(frame, type_5_text, (70,240),cv2.FONT_HERSHEY_SIMPLEX, 0.6, (240,240,240),2)
    
    #add text
    type_6_text='{}: {}%'.format(class_names[5],int(type_6_pred*100))
    cv2.putText(frame, type_6_text, (70,270),cv2.FONT_HERSHEY_SIMPLEX, 0.6, (240,240,240),2)
    
    #showw the frame
    cv2.imshow("Test out", frame)
    
    key=cv2.waitKey(1)
    
    if key & 0xFF == ord("q"):
        break
        
        
camera.release()
cv2.destroyAllWindows()
```
